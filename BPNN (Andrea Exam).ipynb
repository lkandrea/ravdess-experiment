# Import
Bagian ini merupakan bagian untuk import library yang akan digunakan dalam eksperimen

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math

from sklearn.preprocessing import MinMaxScaler
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
from google.colab import drive 
```

# Dataset
Dataset yang akan digunakan akan diupload kedalam Google Drive agar dapat diakses dalam eksperimen

```
drive.mount('/content/gdrive', force_remount=True)

df = pd.read_csv('gdrive/My Drive/Public/slump_test.data')
df = df.drop(['No', 'SP', 'SLUMP(cm)', 'FLOW(cm)'], axis=1)
df
```

# Splitting Data
Bagian ini akan membagi data menjadi training dan test. Pembagian akan dilanjutkan untuk vektor x dan y pada masing masing dataset.

```
x_scaler = MinMaxScaler(feature_range=(0,1))
y_scaler = MinMaxScaler(feature_range=(0,1))

x_dataset = x_scaler.fit_transform(df.values)
y_dataset = y_scaler.fit_transform(df.values[:, 6].reshape(-1,1))

divider_dataset = math.ceil(len(x_dataset) * .8)

x_train = x_dataset[:divider_dataset]
y_train = y_dataset[:divider_dataset]

x_test = x_dataset[divider_dataset:]
y_test = y_dataset[divider_dataset:]
```

# Model 1
Model dengan 1 Hidden Layer 

```
model1 = Sequential()
model1.add(Dense(5, input_dim=7, kernel_initializer='normal', activation='linear'))
model1.compile(optimizer='adam', loss='mse')
model1.fit(x_train, y_train, epochs=100, batch_size=10, verbose=0)

predictions = model1.predict(x_test)
predictions = y_scaler.inverse_transform(predictions)

rmse = np.sqrt(np.mean(predictions-y_test)**2)
print("RMSE Model 1 = ", rmse, "\n")
model1.summary()
```

# Model 2
Model dengan 2 Hidden Layer

```
model2 = Sequential()
model2.add(Dense(10, input_dim=7, kernel_initializer='normal', activation='linear'))
model2.add(Dense(5))
model2.compile(optimizer='adam', loss='mse')
model2.fit(x_train, y_train, epochs=100, batch_size=10, verbose=0)

predictions = model2.predict(x_test)
predictions = y_scaler.inverse_transform(predictions)

rmse = np.sqrt(np.mean(predictions-y_test)**2)
print("RMSE Model 2 = ", rmse, "\n")
model2.summary()
```

# Model 3
Model dengan 3 Hidden Layer

```
model3 = Sequential()
model3.add(Dense(20, input_dim=7, kernel_initializer='normal', activation='linear'))
model3.add(Dense(10))
model3.add(Dense(5))
model3.compile(optimizer='adam', loss='mse')
model3.fit(x_train, y_train, epochs=100, batch_size=10, verbose=0)

predictions = model3.predict(x_test)
predictions = y_scaler.inverse_transform(predictions)

rmse = np.sqrt(np.mean(predictions-y_test)**2)
print("RMSE Model 3 = ", rmse, "\n")
model3.summary()
```
